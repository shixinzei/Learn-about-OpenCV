```python
from imutils import *
```

```python
image = imread('image.jpg')
show(image)
print(image.shape)
```

```python
width = 150
high = 150 
image = cv2.resize(image, (width, high))
show(image)
print(image.shape)
```

```python
image = imread('image.jpg')
width = 80
high = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image, (width, high))
show(image)
print(image.shape)
```

```python
# 5种插值方法：
# 最邻近
image = imread('image.jpg')
width = 150
high = 150 
image = cv2.resize(image, (width, high), interpolation=cv2.INTER_NEAREST)
show(image)
print(image.shape)
```

```python
# 5种插值方法：
# 双线性
image = imread('image.jpg')
width = 150
high = 150 
image = cv2.resize(image, (width, high), interpolation=cv2.INTER_LINEAR)
show(image)
```

```python
# 5种插值方法：
# 基于像素区域
image = imread('image.jpg')
width = 150
high = 150 
image = cv2.resize(image, (width, high), interpolation=cv2.INTER_AREA)
show(image)
print(image.shape)
```

```python
# 5种插值方法：
# 立方插值
image = imread('image.jpg')
width = 150
high = 150 
image = cv2.resize(image, (width, high), interpolation=cv2.INTER_CUBIC)
show(image)
print(image.shape)
```

```python
# 5种插值方法：
# 兰索斯插值
image = imread('image.jpg')
width = 150
high = 150 
image = cv2.resize(image, (width, high), interpolation=cv2.INTER_LANCZOS4)
show(image)
print(image.shape)
```
