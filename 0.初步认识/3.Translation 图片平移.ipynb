```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def imread(image):
    image = cv2.imread(image)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image = imread('image.jpg')
M = np.float32([[1,0,250],[0,1,500]]) # (1,0) 左右 （0,1）上下 250 500 像素
shifted = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
show(shifted)
```

```python
image = imread('image.jpg')
M = np.float32([[1,0,-250],[0,1,-500]])
shifted = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
show(shifted)
```
