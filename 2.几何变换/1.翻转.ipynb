```python
# 水平翻转

image = imread('test.jpg')
image = cv2.flip(image,1)
show(image)
```

```python
# 垂直翻转

image = imread('test.jpg')
image = cv2.flip(image,0)
show(image)
```

```python
# 水平+垂直翻转

image = imread('test.jpg')
image = cv2.flip(image,-1)
show(image)
```
